```python
#get data from web
url = "https://raw.githubusercontent.com/WalePhenomenon/climate_change/master/fuel_ferc1.csv"
```

```python
# read data into notebook and save in variable
raw_fuel_data = pd.read_csv(url, error_bad_lines = False)
```

```python
raw_fuel_data
```

```python
# have a look at the data sample stats
raw_fuel_data.describe(include = "all")
```

```python
# group by fuel unit
raw_fuel_data.groupby("fuel_type_code_pudl")["fuel_type_code_pudl"].count()
```

```python
# check for missing 
raw_fuel_data.isnull().sum()
```

```python
raw_fuel_data[["fuel_unit"]] = raw_fuel_data[["fuel_unit"]].fillna(value = "mcf")
```

```python
#check if missing is filled
raw_fuel_data.isnull().sum()
```

```python
raw_fuel_data.groupby("fuel_type_code_pudl")["fuel_type_code_pudl"].count()
```

```python
# group data by type of fuel
raw_fuel_data.groupby("fuel_type_code_pudl").first()
```

```python
# fuel type lowest ave fuel cost per unit burned
raw_fuel_data.groupby("fuel_type_code_pudl")["fuel_cost_per_unit_burned"].min()
```

```python
#determine std of fuel mmbtu per unit. round off to 5
round (np.std(raw_fuel_data.fuel_mmbtu_per_unit), 5)
```

```python
#determine 75 percentile for fuel mmbtu per unit. round off to 3
round (np.percentile(raw_fuel_data.fuel_mmbtu_per_unit, q = 75),3) 
```

```python
#skewness. round off to 3
print(round(skew(raw_fuel_data.fuel_qty_burned), 3))
```

```python
# kurtosis. round off to 3
print(round(kurtosis(raw_fuel_data.fuel_qty_burned), 3))
```

```python
# feature with most missing values & total
raw_fuel_data.isnull().sum()
```

```python
# % of missing value as ratio of total rows
per_as_a_fac = 180 / len(raw_fuel_data) 
print (round(per_as_a_fac * 100, 4))
```
